##**¿QUE ES ?**

Pandas es una biblioteca para Ciencia de Datos de código abierto (open source), construida sobre el lenguaje Python, y que proporciona un enfoque rápido y flexible, con estructuras robustas para trabajar con datos relacionales (o etiquetados), y todo eso de manera sencilla y intuitiva.

A pesar de que el nombre de la biblioteca está asociado al mamífero de la familia Ursidae, al igual que Python está asociado erróneamente con la especie de cobra, el nombre de la biblioteca Pandas deriva del término Panel Data, un concepto en inglés relacionado con el campo de estudio de la econometría.

De manera general, Pandas puede ser utilizado para diversas actividades y procesos, entre ellos: limpieza y tratamiento de datos, análisis exploratorio de datos (EDA), soporte en actividades de Machine Learning, consultas y queries en bases de datos relacionales, visualizaciones de datos, web scraping y mucho más. Además de eso, también posee una excelente integración con varias otras bibliotecas muy utilizadas en Ciencia de Datos, como: Numpy, Scikit-Learn, Seaborn, Altair, Matplotlib, Plotly, Scipy y otras.




##**¿PARA QUE SIRVE?**

Pandas sirve para:

Organizar datos en tablas.
Leer archivos de diferentes formatos como CSV, Excel o JSON.
Limpiar datos incorrectos o incompletos.
Realizar cálculos y análisis estadísticos.
Filtrar y transformar información.
Preparar datos para modelos de Inteligencia Artificial y Machine Learning.

Gracias a Pandas, los analistas y científicos de datos pueden trabajar con grandes volúmenes de información de manera rápida y eficiente.



##**¿COMO FUNCIONA PANDAS?**
Dentro del paquete Pandas, tenemos dos objetos primarios importantes: las Series y los DataFrames. Y para comprender un poco mejor sobre estas estructuras, vamos a utilizar como ejemplo un conjunto de datos llamado iris, que contiene información sobre características de especies de las flores de Íris.

###**SERIES**
Series
Las Series son objetos de tipo array unidimensional, con un eje de etiquetas, también llamado index, que es responsable de identificar cada registro.

###**DATAFRAME**
DataFrame
Los DataFrames son objetos bidimensionales, de tamaño variable. Su formato es similar a una tabla, donde los datos se organizan en filas y columnas. Además, mientras que podemos pensar en las Series como una única columna, el DataFrame sería una colección de varias Series bajo un mismo índice.

##**OBJETIVO PRINCIPAL**

El objetivo principal de pandas es proporcionar herramientas que permitan trabajar con datos de forma sencilla, rápida y estructurada dentro de Python.

Esta librería ayuda a convertir grandes cantidades de información en datos organizados y fáciles de analizar. Gracias a sus funciones, los usuarios pueden manipular tablas, ordenar registros, filtrar información y generar análisis sin necesidad de escribir procesos complejos.

Pandas también tiene como finalidad optimizar el tratamiento de datos en proyectos tecnológicos y empresariales, permitiendo automatizar tareas que normalmente tomarían mucho tiempo de manera manual.

Otra función importante es servir como base para procesos de análisis avanzado, visualización de información y desarrollo de modelos de Inteligencia Artificial, ya que permite preparar los datos antes de ser utilizados en otras herramientas o algoritmos.

Por su eficiencia y facilidad de uso, Pandas es ampliamente utilizada en áreas como análisis de negocios, ciencia de datos, investigación y machine learning.

##**FUNCIONES PRINCIPALES Y CARACTERISTICAS**

### 1. Lectura y escritura de datos

Permite importar y exportar información desde diferentes formatos de archivos.

Funciones más usadas:

* `read_csv()` → Leer archivos CSV.
* `read_excel()` → Leer archivos Excel.
* `to_csv()` → Guardar datos en CSV.
* `to_excel()` → Guardar datos en Excel.

Ejemplo:

```python
import pandas as pd

df = pd.read_csv("ventas.csv")
```

---

### 2. Manipulación de datos

Ayuda a organizar, modificar y transformar la información.

Funciones principales:

* `head()` → Mostrar las primeras filas.
* `tail()` → Mostrar las últimas filas.
* `columns` → Ver nombres de columnas.
* `rename()` → Cambiar nombres de columnas.
* `drop()` → Eliminar columnas o filas.

Ejemplo:

```python
df.head()
```

---

### 3. Filtrado y selección de información

Permite buscar datos específicos dentro del conjunto de datos.

Funciones y operadores:

* `loc[]`
* `iloc[]`
* Condiciones con `>`, `<`, `==`

Ejemplo:

```python
df[df["Ventas"] > 100]
```

---

### 4. Limpieza de datos

Muy útil para preparar información antes del análisis o modelos de IA.

Funciones importantes:

* `isnull()` → Detectar valores vacíos.
* `dropna()` → Eliminar valores nulos.
* `fillna()` → Reemplazar valores vacíos.
* `duplicated()` → Detectar datos repetidos.

Ejemplo:

```python
df.fillna(0)
```

---

### 5. Análisis y estadísticas

Permite obtener información resumida y estadísticas de los datos.

Funciones más utilizadas:

* `describe()` → Estadísticas generales.
* `mean()` → Promedio.
* `sum()` → Suma total.
* `value_counts()` → Contar valores repetidos.

Ejemplo:

```python
df.describe()
```

---

### 6. Agrupación de datos

Sirve para analizar información por categorías.

Función principal:

* `groupby()`

Ejemplo:

```python
df.groupby("Canal")["Ventas"].sum()
```

---

### 7. Ordenamiento de información

Permite organizar los datos de mayor a menor o viceversa.

Función:

* `sort_values()`

Ejemplo:

```python
df.sort_values("Ventas", ascending=False)
```

---

### 8. Unión y combinación de tablas

Permite trabajar con múltiples archivos o tablas de datos.

Funciones:

* `merge()`
* `concat()`

Ejemplo:

```python
pd.merge(df1, df2, on="ID")
```

---

### 9. Creación y modificación de columnas

Ayuda a generar nuevos datos a partir de otros existentes.

Ejemplo:

```python
df["Total"] = df["Cantidad"] * df["Precio"]
```

---

### 10. Visualización básica de datos

Pandas permite generar gráficos simples para analizar información rápidamente.

Funciones:

* `plot()`
* `hist()`

Ejemplo:

```python
df["Ventas"].plot()
```
